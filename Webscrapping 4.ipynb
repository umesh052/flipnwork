```python
name=[]
des=[]
lang=[]
for i in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]'):
    name.append(i.text)
    
for i in driver.find_elements_by_xpath('//p[@class="col-9 color-text-secondary my-1 pr-4"]'):
    des.append(i.text)

        
```

```python
#fetching all the url
urls=[]
for i in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]//a'):
    urls.append(i.get_attribute('href'))

for i in urls:
    driver.get(i)
    try:
        for i in driver.find_elements_by_xpath('//div[@class="BorderGrid-cell"]//ul[@class="list-style-none"]'):
            lang.append(i.text.replace("\n"," "))
    except NoSuchElementException:
        lang.append("---")
        None
```

```python
github=pd.DataFrame({"Repository title":name,
                    "Repository description":des,
                    " Language used":lang})
github
```
