# Write a python program to scrape the details for all billionaires from www.forbes.com. 
Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, 
“Industry”

```python
driver.get("https://www.forbes.com/?sh=3ac8c6bc2254")
```

```python
for i in driver.find_elements_by_xpath('//li[@data-title="Billionaires"]//a'):
    urls.append(i.get_attribute('href'))
```

```python
rank=[]
name=[]
net=[]
age=[]
country=[]
source=[]
industry=[]
```

```python


while True:
    try:
        for j in driver.find_elements_by_xpath('//div[@class="rank"]'):
            rank.append(j.text.replace(".",""))
    
        for j in driver.find_elements_by_xpath('//div[@class="personName"]'):
            name.append(j.text)
    
        for j in driver.find_elements_by_xpath('//div[@class="netWorth"]'):
            net.append(j.text)
    
        for j in driver.find_elements_by_xpath('//div[@class="age"]'):
            age.append(j.text)
    
        for j in driver.find_elements_by_xpath('//div[@class="countryOfCitizenship"]'):
            country.append(j.text)
    
        for j in driver.find_elements_by_xpath('//div[@class="source"]'):
            source.append(j.text)
    
        for j in driver.find_elements_by_xpath('//div[@class="category"]'):
            industry.append(j.text) 
            
        next=driver.find_element_by_xpath('//div[@class="next-page"]')
        next.click()
        
        
    except:
        print("no more pages")
        break
```

```python
df=pd.DataFrame({"RANK":rank,
                 "NAME":name,
                 "NET_WORTH":net,
                 "AGE":age,
                 "COUNTRY/TERRITORY":country,
                 "SOURCE":source,
                 "INDUSTRY":industry
})
df
```
